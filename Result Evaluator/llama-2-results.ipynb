# Llama-2 Set 1

Learning Rate

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter

# Increase font sizes globally
plt.rcParams.update({'font.size': 13,  # Adjust base font size as needed
                     'axes.titlesize': 18,  # Adjust for title
                     'axes.labelsize': 16,  # Adjust for x and y labels
                     'xtick.labelsize': 14,  # Adjust for x-axis tick labels
                     'ytick.labelsize': 14,  # Adjust for y-axis tick labels
                     'legend.fontsize': 14})  # Adjust for legend

# Function to format the y-axis ticks
def format_tick(x, pos):
    return f'{x*1e4:.2f}'

# Load the data
data = pd.read_csv('Llama-2/Set-1/learning-rate.csv')

# Set the style with white background
plt.style.use('seaborn-whitegrid')

# Create a color palette
palette = plt.get_cmap('Set1')

# Plot data
plt.figure(figsize=(10, 6))
plt.plot(data['Epochs'], data['learning-rate'], marker='', color='blue', linewidth=2.5, alpha=0.9, label='Learning Rate')

# Add titles and labels
plt.title('Learning Rate over Epochs', loc='center', fontweight=0, color='black', pad=10)  # Font size adjusted here as well, for consistency
plt.xlabel('Epochs', labelpad=8)
plt.xticks(np.arange(0, 21, 1))
plt.yticks(np.arange(0, 0.0006, 0.0001))
plt.ylabel('Learning Rate in $10^{-4}$', labelpad=8)

# Set grid with added opacity
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.75)

# Format y-axis
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_tick))

# Show legend
plt.legend()

# Save the figure
plt.savefig('learning_rate_vs_epochs.svg', dpi=300)

# Show plot
plt.show()
```
