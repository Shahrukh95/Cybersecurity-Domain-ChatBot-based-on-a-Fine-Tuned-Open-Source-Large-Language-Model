```python
import json
import os
import pandas as pd
from openai import OpenAI
from dotenv import load_dotenv
import time
import glob

# Load environment variables
load_dotenv()

def read_excel_file(file_path):
    return pd.read_excel(file_path, usecols=[0, 1])

def clean_text(text):
    return text.replace('\n', ' ').replace('\r', ' ').strip()

def get_rephrased_qa(question, answer, client):
    prompt = f"The following is a pair of Question and Answer that you have to rephrase completely. Question: '''{question}'''. Answer: '''{answer}'''. The response should be in JSON format. Each set of Question-Answer pairs should be an object inside an array, and their key and value should be called 'question' and 'answer' respectively. The parent key should be called 'data'."

    try:
        completion = client.chat.completions.create(
            model="gpt-4-1106-preview",
            messages=[{'role': 'user', 'content': prompt}],
            response_format={"type": "json_object"}
        )
        response = completion.choices[0].message.content.strip()
        print(f"Tokens used in GPT-4-Turbo Response: {completion.usage.total_tokens}")
        return response
    except Exception as e:
        print(f"Error in GPT-4-Turbo API call: {e}")
        return None

def append_to_json(json_data, filename):
    if os.path.isfile(filename):
        with open(filename, 'r+') as file:
            file_data = json.load(file)
            file_data.append(json_data)
            file.seek(0)
            json.dump(file_data, file, indent=4)
    else:
        with open(filename, 'w') as file:
            json.dump([json_data], file, indent=4)

def append_to_excel(qa_pair, filename):
    df = pd.DataFrame([qa_pair], columns=['Question', 'Answer'])
    if os.path.isfile(filename):
        book = pd.read_excel(filename)
        df = pd.concat([book, df], ignore_index=True)
    df.to_excel(filename, index=False)

def process_files(folder_path, output_json, output_excel):
    client = OpenAI(api_key=os.getenv("API_KEY"))
    files = glob.glob(os.path.join(folder_path, "*.xlsx"))

    for file in files:
        print(f"\nProcessing file: {file}")
        data = read_excel_file(file)
        for index, row in data.iterrows():
            print(f"\nProcessing row: {index + 1}")
            question, answer = clean_text(row[0]), clean_text(row[1])
            json_response = get_rephrased_qa(question, answer, client)
            if json_response:
                try:
                    response_data = json.loads(json_response)
                    for qa_pair in response_data.get("data", []):
                        append_to_json(qa_pair, output_json)
                        append_to_excel((qa_pair.get('question'), qa_pair.get('answer')), output_excel)
                except json.JSONDecodeError as e:
                    print(f"JSON parsing error: {e}")
                    print(f"Invalid JSON response: {json_response}")
            print("Taking a break for 3 seconds")
            time.sleep(3)

if __name__ == "__main__":
    output_folder_path = 'Output'
    output_json_path = 'rephrased_data.json'
    output_excel_path = 'rephrased_data.xlsx'
    process_files(output_folder_path, output_json_path, output_excel_path)
```
